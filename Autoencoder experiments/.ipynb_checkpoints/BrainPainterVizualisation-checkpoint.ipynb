```python
import pandas as pd
```

```python
# Mean woman / man brains or age brains 
data = pd.read_csv("../data/brains_age.csv")
```

```python
data
```

```python
template = pd.read_csv("../data/DK_template.csv")
```

```python
data['Image-name-unique'] = ['Male', 'Female']
# to reproduce for age
# data['Image-name-unique'] = ['22-25', '26-30', '31+']
data = data.set_index('Image-name-unique').reset_index()
```

```python
# columns matching

dk_data = pd.DataFrame(columns=template.columns)
for col in dk_data.columns:
    if col == 'Image-name-unique':
        dk_data[col] = data[col]
        continue
    if 'Right' in col:
        name = 'fs_r_' + col.split('-')[1].lower() + '_thck'
    elif 'Left' in col:
        name = 'fs_l_' + col.split('-')[1].lower() + '_thck'
    else:
        d_type = 'vol'
        if col == 'Accumbens-area':
            name = f'fs_l_accumbensarea_{d_type}'
        elif col == 'Inf-Lat-Vent':
            name = f'fs_l_inflatvent_{d_type}'
        elif col == 'Cerebellum-White-Matter':
            name = f'fs_l_cerebellum_wm_{d_type}'
        elif col == 'Thalamus-Proper':
            name = f'fs_l_thalamusproper_{d_type}'
        elif col == 'Cerebellum-Cortex':
            name = f'fs_l_cerebellum_cort_{d_type}'
        elif col == 'Hippocampus':
            name = f'fs_l_hippo_{d_type}'
        elif col == 'Lateral-Ventricle':
            name = f'fs_l_latvent_{d_type}'
        elif col == 'VentralDC':
            name = f'fs_l_ventdc_{d_type}'
        else:
            name = col.lower() + f'_{d_type}'
    candidates = list(filter(lambda x: name in x.lower(), data.columns))
    if len(candidates) == 0:
        print(f"!!! Not found: {col}, {name}")
        dk_data[col] = [0.5, 0.5, 0.5]
    else:
        print(f"Found candidate: {candidates[0]}")
        dk_data[col] = data[candidates[0]]
#         if 'vol' in name:
#             dk_data[col] = dk_data[col]/1000
```

```python
dk_data.to_csv("processed_brains_age.csv", index=False)

# Afterwards, load processed brains, and use https://brainpainter.csail.mit.edu/ to generate vizualisations
```
